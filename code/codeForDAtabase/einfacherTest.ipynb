```python
from geopy.geocoders import Nominatim
from concurrent.futures import ThreadPoolExecutor
geolocator = Nominatim(user_agent="geoapiExercises")

# Create a list of locations to geocode
locations = df['Location'].tolist()

# Set the number of threads to use
num_threads = 10

# Create a ThreadPoolExecutor
with ThreadPoolExecutor(max_workers=num_threads) as executor:
    # Geocode the locations in batches
    batch_size = len(locations) // num_threads
    for i in range(0, len(locations), batch_size):
        batch = locations[i:i+batch_size]
        geo_results = executor.map(geolocator.geocode, batch)
        for j, geo in enumerate(geo_results):
            if geo is not None:
                latitude = geo.latitude
                longitude = geo.longitude
                df.at[i+j, 'Latitude'] = latitude
                df.at[i+j, 'Longitude'] = longitude
            else:
                print(f"Error geocoding {batch[j]}")
```

```python
from geopy.exc import GeocoderTimedOut
def get_lat_long(location):
    try:
        geolocator = Nominatim(user_agent="Geopy Library")
        location = geolocator.geocode(location)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except GeocoderTimedOut:
        print("Geocoding service timed out.")
        return None, None
```

```python
df['Latitude'], df['Longitude'] = zip(*df['Location'].apply(get_lat_long))
```
